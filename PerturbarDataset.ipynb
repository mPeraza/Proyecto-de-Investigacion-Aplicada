```python

data =  raw_data.drop(['usage_counter','normal_prio', 'policy', 'vm_pgoff', 'task_size','cached_hole_size', 'hiwater_rss', 'nr_ptes', 'min_flt','lock', 'cgtime', 'signal_nvcsw'], axis=1)
```

```python
import random
from numpy.ma.core import std
def crear_perturbacion(alfa, columna_a_perturbar, cant_datos_pertubar):
  desviacion_estandar = data[columna_a_perturbar].std()
  cambio = alfa * desviacion_estandar
  indice=-1
  
  for i in range(cant_datos_pertubar):
    indice = random.randint(0,50000)
    #indice = 1
    print("Indice: ",indice)
    valor_inicial = data[columna_a_perturbar][indice]
    print("Valor inicial",valor_inicial)
    valor_modificado = valor_inicial + cambio  
    data.at[indice, columna_a_perturbar]= valor_modificado
    print("Desviacion estandar", desviacion_estandar)
    print("Valor modificado",valor_modificado)
```

```python
def seleccionar_columnas(cant_columnas):
  columnas_a_perturbar = ["prio", "static_prio", "vm_truncate_count", "mm_users", "map_count", "total_vm", "shared_vm", "exec_vm", "reserved_vm", "end_data", "last_interval", "nvcsw", "nivcsw", "maf_fit", "fs_excl_counter", "utime", "stime", "gtime"]
  columnas = []
  for i in range(cant_columnas):
    columna_a_perturbar = np.random.choice(columnas_a_perturbar, replace=False)
    columnas.append(columna_a_perturbar)
  return columnas
  
```

```python
def perturbar_columnas(cant_columnas, alfa, cant_datos_pertubar):
  columnas_a_perturbar = seleccionar_columnas(cant_columnas)
  for i in range(cant_columnas):
    print (columnas_a_perturbar[i])
    crear_perturbacion(alfa, columnas_a_perturbar[i], cant_datos_pertubar)
```
